```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline

## GridSearchCV
from sklearn.model_selection import GridSearchCV
```

```python
df = pd.read_csv("boston_data.csv", index_col=0)
df.head(2)
```

```python
x = df.drop("medv", axis=1)
y = df["medv"]
```

```python
std_scaler = StandardScaler()
knn_model = KNeighborsRegressor(n_neighbors=5)


pipeline_boston = Pipeline(
        [("scale", std_scaler),
        ("model", knn_model)]
    )
pipeline_boston
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
pipeline_boston.fit(x_train, y_train)
```

```python
pipeline_boston.score(x_test, y_test)
```

```python
std_scaler = StandardScaler()
knn_model = KNeighborsRegressor(n_neighbors=5)


pipeline_boston = Pipeline(
        [("scale", std_scaler),
        ("model", knn_model)]
    )
pipeline_boston
```

```python
neighbors_range = list(range(2, 11))
neighbors_range
```

```python
"model" + "__" + "n_neighbors"
```

```python

grid_model = GridSearchCV(estimator= pipeline_boston, 
             param_grid= {
                 'model__n_neighbors' : neighbors_range
             })

grid_model
```

```python
grid_model.fit(x_train, y_train)
```

```python
results_df = pd.DataFrame(grid_model.cv_results_)
results_df[["param_model__n_neighbors", "rank_test_score"]]
```

```python
results_df[["param_model__n_neighbors", "rank_test_score"]].sort_values("rank_test_score")
```
