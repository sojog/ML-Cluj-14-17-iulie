```python
from sklearn.datasets import load_diabetes
```

```python
diabetes_dataset = load_diabetes()
diabetes_dataset.keys()
```

```python
x, y = load_diabetes(return_X_y=True)
```

```python
# x - features
"samples:", x.shape[0], "featues", x.shape[1], y.shape
```

```python
# y - target
y.min(), y.max()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
std_scaler = StandardScaler()
std_scaler
```

```python
x_scaled = std_scaler.fit_transform(x, y)
x_scaled.min(), x_scaled.max()
```

```python
x.min(), x.max()
```

```python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.pipeline import Pipeline

std_scaler = StandardScaler()
knn_model = KNeighborsRegressor(n_neighbors=5)


pipeline_diabetes = Pipeline(
        [("scale", std_scaler),
        ("model", knn_model)]
    )
pipeline_diabetes
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(pipeline_diabetes, param_grid= {
    "model__n_neighbors" : list(range(3, 7))
})
```

```python
grid_model.fit(x, y)
```

```python
grid_model.best_estimator_
```

```python
grid_model.best_score_
```
