```python
from scipy import stats
```

```python
x, y = make_regression(n_samples=30, n_features=1)

rezultat = stats.linregress(x.flatten(), y)
def scipy_func(x):
    return rezultat.slope * x  - rezultat.intercept

plt.scatter(x, y);
plt.plot(x, y);

plt.plot(x, list(map(scipy_func, x)));
```

```python
rezultat = stats.linregress(x.flatten(), y)
def scipy_func(x):
    return rezultat.slope * x  + rezultat.intercept

plt.scatter(x, y);
plt.plot(x, y);

plt.plot(x, list(map(scipy_func, x)), c="orange");


def sklearn_func(x):
    return model.coef_ * x + model.intercept_

plt.plot(x, list(map(sklearn_func, x)), c="black");
```

```python
x, y = make_regression(n_features=1, n_samples=20, noise=2)

rezultat = stats.linregress(x.flatten(), y)
def scipy_func(x):
    return rezultat.slope * x  + rezultat.intercept

plt.scatter(x, y);
# plt.plot(x, y);

plt.plot(x, list(map(scipy_func, x)), c="orange");

model = LinearRegression()
model.fit(x, y)

def sklearn_func(x):
    return model.coef_ * x + model.intercept_

plt.plot(x, list(map(sklearn_func, x)), c="black");
```
