```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("boston_data.csv", index_col=0)
df.head()
```

```python
y = df["medv"]
type(y), len(y)
```

```python
x = df.drop("medv", axis=1)
type(x), x.shape
```

```python
model = LinearRegression()
model
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model.fit(x_train, y_train)
```

```python
y_prezis = model.predict(x_test)
y_prezis
```

```python
y_prezis.shape, y_test.shape
```

```python
comparatie_df = pd.DataFrame({
    "y_prezis" : y_prezis,
    "y_test": y_test
})
comparatie_df
```

# KNN

```python
from sklearn.neighbors import KNeighborsRegressor
```

```python
knn_model = KNeighborsRegressor(n_neighbors=2)
knn_model
```

```python
knn_model.fit(x_train, y_train)
```

```python
knn_results = knn_model.predict(x_test)
```

```python
comparatie_df = pd.DataFrame({
    "y_test": y_test,
    "y_prezis_linear" : y_prezis,
    "y_prezis_knn":knn_results
  
})
comparatie_df
```

```python
lm_score = model.score(x_test, y_test)
knn_score = knn_model.score(x_test, y_test)
```
