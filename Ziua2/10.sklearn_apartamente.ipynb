```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
apartamente = [(37, 65), (44, 70), (55, 80), (80, 120), (60, 95), (92, 205), (41, 75), (170, 400)]

suprafete = list(map(lambda x:x[0], apartamente))
suprafete


preturi = [ i[1] for i in apartamente ]
preturi
```

```python
plt.scatter(suprafete, preturi);
```

```python
model = LinearRegression()
model
```

```python
suprafete = np.array(suprafete)
suprafete = suprafete.reshape((-1, 1))
suprafete
```

```python
model.fit(suprafete, preturi)
```

```python

plt.scatter(suprafete, preturi);
plt.plot(suprafete, list(map(lambda x: model.coef_ * x + model.intercept_,suprafete)));
```
